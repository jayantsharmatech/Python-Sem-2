```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Data Loading
df = pd.read_csv('PS09_Dataset_9.csv')

# 2. Preprocessing: Fill missing values with the average of that column
df.fillna(df.mean(), inplace=True)

print("Cleaned Dataset (ALL ROWS):")
# Using to_string() instead of head() forces it to print the entire dataset
print(df.to_string())

# 3. Calculations (NumPy): Add columns A, B, C, and D together for each row
df['Total_Score'] = np.sum(df[['A', 'B', 'C', 'D']], axis=1)

# Calculate the average of each column to use in the bar chart
column_averages = np.mean(df[['A', 'B', 'C', 'D']], axis=0)

print("\nBasic Calculations:")
print(f"Overall average of all total scores: {np.mean(df['Total_Score'])}")

# 4. Visualization 1: Simple Bar Chart
plt.bar(['A', 'B', 'C', 'D'], column_averages)
plt.title('Average Value of Each Column')
plt.xlabel('Columns')
plt.ylabel('Average')
plt.show()

# 5. Visualization 2: Simple Histogram
plt.hist(df['Total_Score'])
plt.title('Spread of Total Scores')
plt.xlabel('Total Score')
plt.ylabel('Frequency')
plt.show()

# 6. Summary of Insights
print("\n Summary of Insights ")
print("1. There were no major outliers; the data was cleaned by replacing missing values with column averages.")
print("2. The Bar Chart shows that columns A, B, C, and D all have very similar average values.")
print("3. The Histogram shows that most row totals cluster around the center of the distribution.")
```
